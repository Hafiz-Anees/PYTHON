```python
import random

# Countries & informative hints
country_name = [
    'pakistan', 'india', 'china', 'japan', 'nepal',
    'afghanistan', 'bangladesh', 'sri lanka', 'iran', 'turkey',
    'france', 'germany', 'italy', 'canada', 'brazil'
]

hints = [
    'Capital: Islamabad, Famous for Karakoram mountains',
    'Capital: New Delhi, Famous for Taj Mahal',
    'Capital: Beijing, Known as the world’s factory',
    'Capital: Tokyo, Land of the Rising Sun',
    'Capital: Kathmandu, Home of Mount Everest',
    'Capital: Kabul, Known as the “Heart of Asia”',
    'Capital: Dhaka, Famous for jute and rivers',
    'Capital: Colombo, Famous for tea and beaches',
    'Capital: Tehran, Known for Persian culture',
    'Capital: Ankara, Famous for Bosphorus strait',
    'Capital: Paris, City of Love & Eiffel Tower',
    'Capital: Berlin, Known for Berlin Wall history',
    'Capital: Rome, Famous for Colosseum & Vatican',
    'Capital: Ottawa, Famous for maple leaves',
    'Capital: Brasília, Famous for Amazon rainforest'
]


lifes = 5 

def GenerateRandomCountry():
    # Pick random country
    randomIndex = random.randint(0, len(country_name) - 1)
    random_country = country_name[randomIndex]
    random_hint = hints[randomIndex]
    return random_country, random_hint
    
    
def EncryptedCountry(random_country):
    hidden_word = ["*"] * len(random_country)
    return hidden_word

def GuessTheCountry():
    print("Welcome to Guess The Country Game!")
    print("You have 5 lives to guess the country name.")
    print("Let's begin!")
    random_country, random_hint = GenerateRandomCountry()
    hidden_country = EncryptedCountry(random_country)
    print("🎮 Guess the Country Name!")
    print("Hint:", random_hint)
    print("Word:", "".join(hidden_country))
    
    while "*" in hidden_country:
        guess = input("Enter a character: ").lower()
        if len(guess) != 1:
            print("❌ Please enter only one character.")
            continue

        correct = False
        for i in range(len(random_country)):
            if guess == random_country[i]:
                hidden_country[i] = guess
                print(hidden_country)
                correct = True

        if correct:
            print("✅ Correct guess!")
        else:
            print("❌ Wrong guess!")
            global lifes
            lifes -= 1
            if lifes == 0:
                return random_country
                
    return random_country


# code start here
if __name__ == "__main__":
    ans = GuessTheCountry()
    if lifes == 0:
        print("💀 You've run out of lives! The country was:", ans)
    else:
        print("🎉 Congratulations! You've guessed the country:", ans)
```
